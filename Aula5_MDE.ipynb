```python
# Remover instâncias duplicadas
df = df.drop_duplicates()

# Selecionar características para o modelo
features = df.columns.drop(['G2', 'G3'])  # Exclui G2 e G3 das características
X_possiveis = df[features]
y = df['G3']  # Variável de saída

# Verificar a nova estrutura dos dados após remoção de duplicatas e seleção de características
X_possiveis.head(), y.head()
```

```python
# Identificar colunas categóricas
colunas_categoricas = X_possiveis.select_dtypes(include=['object']).columns

# Aplicar one-hot encoding às variáveis categóricas
X_encoded = pd.get_dummies(X_possiveis, columns=colunas_categoricas)

# Visualizar as primeiras linhas do dataframe após o one-hot encoding
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Normalizar os dados
scaler = MinMaxScaler()
scaler.fit(X_train)  # Ajusta o scaler apenas nos dados de treinamento

X_train_scaled = scaler.transform(X_train)  # Aplica a transformação nos dados de treinamento
X_test_scaled = scaler.transform(X_test)  # Aplica a mesma transformação nos dados de teste

# Verificar as primeiras linhas dos dados de treinamento normalizados para garantir que a normalização foi aplicada
X_train_scaled[:5]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Inicializar os modelos
model_linear = LinearRegression()
model_tree = DecisionTreeRegressor(random_state=42)

# Treinar os modelos
model_linear.fit(X_train_scaled, y_train)
model_tree.fit(X_train_scaled, y_train)

# Fazer previsões nos conjuntos de teste
y_pred_linear = model_linear.predict(X_test_scaled)
y_pred_tree = model_tree.predict(X_test_scaled)

# Avaliar os modelos
mse_linear = mean_squared_error(y_test, y_pred_linear)
mse_tree = mean_squared_error(y_test, y_pred_tree)

r2_linear = r2_score(y_test, y_pred_linear)
r2_tree = r2_score(y_test, y_pred_tree)

mse_linear, mse_tree, r2_linear, r2_tree
```
